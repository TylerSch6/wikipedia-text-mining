Import libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Fetch the Wikipedia page
url = "https://en.wikipedia.org/wiki/DeepSeek_(chatbot)"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
print (soup)
```

Parsing

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
title = soup.find("h1").text
print("Page Title:", title)
```

```python
intro_div = soup.find("div", class_="mw-parser-output")
intro_paragraphs = []
if intro_div:
    for tag in intro_div.find_all("p", recursive=False):
        text = tag.get_text(strip=True)
        if text:
            intro_paragraphs.append(text)
        if len(intro_paragraphs) >= 2:
            break
introduction = "\n\n".join(intro_paragraphs) if intro_paragraphs else "No Introduction"
print("\nIntroduction:\n", introduction)
```

 Extract Section Titles and First Paragraphs

```python
sections = soup.find_all("h2")

for section in sections:
    section_title = section.text.strip().replace("[edit]", "")
    section_content_tag = section.find_next("p")
    if section_content_tag:
        section_content = section_content_tag.text.strip()
        print(f"Section Title: {section_title}")
        print(f"Section Content: {section_content[:300]}...")
        print()
```

Fetch the section History

```python
headers = soup.find_all(['h2', 'h3'])

# Iterate through headers to find the "History" section
history_section_found = False
for header in headers:
    if header.get_text(strip=True).lower() == 'history':
        history_section_found = True
        # Start extracting paragraphs after the "History" header
        next_node = header.find_next()
        while next_node and next_node.name not in ['h2', 'h3']:
            if next_node.name == 'p':
                print(next_node.get_text())
            next_node = next_node.find_next()

if not history_section_found:
    print("History section not found.")
```

Extract the links

```python
external_links = soup.find_all("a", {"class": "external text"})
print("External Links:")
for link in external_links:
    href = link.get("href")
    if href:
        print(href)
```

References

```python
references = []
for ref in soup.find_all("cite"):
    link = ref.find("a", href=True)
    if link:
        references.append(link['href'])
print("\nReferences:")
for ref in references:
    print(ref)
```

Sentiment analysis 

```python
from textblob import TextBlob
```

```python
# Get the page text
page_text = soup.get_text()

# Step 3: Clean up the extracted text (optional)
# Remove extra newlines and spaces
page_text = ' '.join(page_text.split())

# Step 4: Perform Sentiment Analysis with TextBlob
blob = TextBlob(page_text)

# Step 5: Output the sentiment analysis results
print("Sentiment Analysis Result:")
print(f"Polarity: {blob.sentiment.polarity}")  # Range: -1 (negative) to 1 (positive)
print(f"Subjectivity: {blob.sentiment.subjectivity}")  # Range: 0 (objective) to 1 (subjective)
```

Polarity: 0.097 — The sentiment of the entire article is slightly positive, but not very strong. A polarity close to 0 means it's fairly neutral, with a slight tendency toward positive sentiment.

Subjectivity: 0.377 — The article is somewhat objective (close to 0), with some level of subjectivity, as expected for a Wikipedia page. This indicates that the article contains a mix of factual content and potentially opinion-based information.
